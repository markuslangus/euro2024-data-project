### Datentypanpassung

Um eine präzisere Handhabung von Datum und Zeit zu ermöglichen, wird die Spalte `date_time` in ein Datetime-Objekt umgewandelt. Dies erleichtert die Handhabung und ermöglicht, datums- und zeitbezogene Funktionen effizienter zu nutzen. Hierfür soll beispielhaft das Addieren der Zeitzonenverschiebung aus der Spalte `utc_offset_hours` zur Spalte `date_time` gezeigt werden.

Da die Spalten `id_match` und `stadium_id` keine logischen Zahlen, sondern vielmehr eindeutige Kennungen (IDs) darstellen, werden diese beiden Spalten in den String-Datentyp umgewandelt. Dadurch wird die korrekte Behandlung dieser Daten sichergestellt.


```python
# Umwandeln der Spalten 'date_time' in ein Datetime-Objekt
df["date_time"] = pd.to_datetime(df["date_time"])

# UTC offset zur date_time Spalte hinzufügen
df['date_time'] = df['date_time'] + pd.to_timedelta(df['utc_offset_hours'], unit='h')

# Umwandeln der Spalten 'id_match' und 'stadium_id' in den Objekt-Datentyp, da es sich um IDs handelt
df["id_match"] = df["id_match"].astype(str)
df["stadium_id"] = df["stadium_id"].astype(str)

# 
columns_to_convert = ['home_score', 'away_score', 'home_score_total', 'away_score_total', 'match_attendance', 'stadium_capacity']
for col in columns_to_convert:
    df[col] = df[col].astype(int)

# Überprüfen der Datentypen nach der Umwandlung
print("Datentyp der Spalte 'date_time' nach der Umwandlung: ", df["date_time"].dtype)    # Erwarteter Output: datetime64
print("Datentyp der Spalte 'id_match' nach der Umwandlung: ", df["id_match"].dtype)      # Erwarteter Output: object
print("Datentyp der Spalte 'stadium_id' nach der Umwandlung: ", df["stadium_id"].dtype)  # Erwarteter Output: object
```

### Entfernung irrelevanter Spalten

Die Spalten `year` und `date` enthalten das Jahr, in dem ein Spiel stattfand, und das jeweilige Kalenderdatum. Da sowohl das Jahr als auch die Informationen über das Kalenderdatum bereits in der Spalte `date_time` enthalten sind, bieten diese Spalten keine zusätzlichen nützlichen Informationen und können entfernt werden. 

Die Spalten `condition_humidity`, `condition_pitch`, `condition_temperature`, `condition_weather` und `condition_wind_speed` enthalten Wetterdaten, die nur für zwei Spiele verfügbar sind. Da diese Informationen zu spärlich sind, um aussagekräftige Analysen zu ermöglichen, werden sie aus dem DataFrame entfernt.

Zusätzlich werden die Spalten `stadium_name_media`, `stadium_name_official`, `stadium_name_event` und `stadium_name_sponsor` entfernt. Diese bieten keine zusätzlichen Erkenntnisse, da die relevante Information bereits durch die Spalte `stadium_name` abgedeckt wird, die als zentrale Quelle für Stadiennamen dient.

Ebenfalls werden die Spalten `stadium_pitch_length` und `stadium_pitch_width` entfernt, da sie identische Werte enthalten. Die Spielfelder bei einem Euro-Turnier sind genormt, sodass diese Informationen redundant sind und keinen Mehrwert für die Analyse bieten.

Dazu können die Spalten `stadium_country_code` und `status` entfernt werden, da alle Spiele des Turniers in Deutschland stattgefunden haben und das Turnier bereits beendet ist.


```python
# Definieren der zu löschenden Spalten
columns_to_drop = [
    "year",                     # Jahr des Spiels, redundant durch Spalte 'date_time'
    "date",                     # Kalenderdatum, redundant durch Spalte 'date_time'
    "utc_offset_hours",         # Zeitzonenverschiebung, bereits in date_time berücksichtigt
    "condition_humidity",       # Wetterdaten, nur für zwei Spiele verfügbar
    "condition_pitch",          # Wetterdaten, nur für zwei Spiele verfügbar
    "condition_temperature",    # Wetterdaten, nur für zwei Spiele verfügbar
    "condition_weather",        # Wetterdaten, nur für zwei Spiele verfügbar
    "condition_wind_speed",     # Wetterdaten, nur für zwei Spiele verfügbar
    "stadium_name_media",       # Redundante Stadionnamen, bereits durch 'stadium_name' abgedeckt
    "stadium_name_official",    # Redundante Stadionnamen, bereits durch 'stadium_name' abgedeckt
    "stadium_name_event",       # Redundante Stadionnamen, bereits durch 'stadium_name' abgedeckt
    "stadium_name_sponsor",     # Redundante Stadionnamen, bereits durch 'stadium_name' abgedeckt
    "stadium_pitch_length",     # Redundante Spielfeldlänge, bei Euro-Turnieren genormt
    "stadium_pitch_width",      # Redundante Spielfeldbreite, bei Euro-Turnieren genormt
    "stadium_country_code",     # Redundante Länderkennung, da alle Spiele im gleichen Land stattfinden
    "status",                   # Redundanter Spielstatus, da das Turnier bereits zu Ende ist
]

# Entfernen der definierten Spalten aus dem DataFrame
df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

## Datenvisualisierung


In diesem Abschnitt werden Visualisierungen präsentiert: Ein Histogramm zur Verteilung der Tordifferenzen aus Sicht des Heimteams, das Einblicke in den Ausgang der Spiele bietet, und eine interaktive Karte, die die geografische Lage der Stadien sowie detaillierte Informationen zu den einzelnen Spielen der UEFA Euro 2024 veranschaulicht.

### Tordifferenzanalyse


Dieses Histogramm zeigt die Verteilung der Tordifferenzen aus Sicht des Heimteams bei der UEFA Euro 2024. Positive Werte bedeuten, dass das Heimteam mehr Tore erzielt hat als das Auswärtsteam, während negative Werte das Gegenteil anzeigen. Die x-Achse ist entsprechend beschriftet, um die Unterschiede klar darzustellen.

```python
# Berechne die Tordifferenz aus Sicht des Heimteams (Tore Heim - Tore Auswärts)
df['goal_difference'] = df['home_score'] - df['away_score']

# Setze den Stil für die Visualisierungen auf 'whitegrid' für bessere Lesbarkeit
sns.set(style="whitegrid")

# Definiere die Bins für das Histogramm, um die Tordifferenzen korrekt darzustellen
bins = np.arange(df['goal_difference'].min() - 0.5, 
                 df['goal_difference'].max() + 1.5, 1)

# Erstelle ein Histogramm für die Tordifferenz aus Sicht des Heimteams
plt.figure(figsize=(8, 5))

# Plotten des Histogramms der Tordifferenz ohne KDE-Kurve
sns.histplot(df['goal_difference'], bins=bins, 
             kde=False, color='blue', edgecolor='black')

# Füge Titel und Achsenbeschriftungen hinzu
plt.title('Verteilung der Tordifferenz')
plt.xlabel('Tordifferenz (Heim minus Auswärts) nach regulärer Spielzeit')
plt.ylabel('Anzahl der Spiele')

# Setze die x-Achse auf ganzzahlige Tordifferenzen
xticks = np.arange(df['goal_difference'].min(), df['goal_difference'].max() + 1, 1)
xtick_labels = [f'+{int(x)}' if x > 0 else str(int(x)) for x in xticks]
plt.xticks(xticks, labels=xtick_labels)

# Setze die y-Achse in 2er-Schritten
yticks = np.arange(0, df['goal_difference'].value_counts().max() + 2, 2)
plt.yticks(yticks)

# Zeige ein Gitternetz für bessere Lesbarkeit an
plt.grid(True)

# Zeige das Diagramm an
plt.show()
```

Das Histogramm zeigt die Verteilung der Tordifferenz aus der Sicht des Heimteams über alle Spiele hinweg. Bei der Ermittlung der Tordifferenz werden die Tore der Heimmannschaft und die Tore der Auswärtsmannschaft voneinander subtrahiert. Gewertet werden alle in der regulären Spielzeit erzielten Tore (Spielstand nach 90 min).

- **Symmetrische Verteilung**: Die Tordifferenz scheint annähernd symmetrisch um den Wert 0 verteilt zu sein, was darauf hindeutet, dass die Spiele relativ ausgeglichen waren. Die meisten Spiele endeten mit einer geringen Tordifferenz (-1, 0, +1).
  
- **Häufigkeit von Unentschieden**: Der Peak bei 0 zeigt, dass der häufigste Spielausgang ein Unentschieden war. In diesen Spielen erzielten die Heim- und Auswärtsmannschaften gleich viele Tore.

- **Positive Tordifferenzen**: Spiele, bei denen das Heimteam mehr Tore erzielte als das Auswärtsteam (positive Tordifferenzen), kommen häufiger vor als Spiele, in denen das Auswärtsteam dominierte (negative Tordifferenzen).

- **Verteilung**: Extreme Tordifferenzen (z.B. -3 oder +4) sind relativ selten, was zeigt, dass die meisten Spiele eng umkämpft waren.

Die Spiele des Uefa Euro 2024 Turniers waren tendenziell ausgeglichen, wobei die Heimteams leicht im Vorteil waren.

```python
# Initialize the map
map_stadiums = folium.Map(location=[51.1657, 10.4515], zoom_start=6)

# For each launch site, add a Circle object based on its coordinate (Lat, Long) values. In addition, add Launch site name as a popup label
for index, row in df.iterrows():
    circle = folium.Circle(
        [row["stadium_latitude"], row["stadium_longitude"]],
        radius=2000,
        color="#d35400",
        fill=True,
    ).add_child(
        folium.Popup(
            f'Official Stadium Name:<br>{row["stadium_name"]}<br><br>Stadium Capacity:<br>{row["stadium_capacity"]}',
            max_width=200,
        )
    )

    marker = folium.map.Marker(
        [row["stadium_latitude"], row["stadium_longitude"]],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % row["stadium_name"],
        ),
    )
    map_stadiums.add_child(circle)
    map_stadiums.add_child(marker)

marker_cluster = MarkerCluster()

def assign_marker_color(matchday):
    if matchday in ["MD1", "MD2", "MD3"]:
        return "lightblue"
    elif matchday == "MD4":
        return "cadetblue"
    elif matchday == "MD5":
        return "blue"
    elif matchday == "MD6":
        return "darkblue"
    else:
        return "darkpurple"


def assign_correct_matchday_names(matchday):
    matchday_mapping = {
        "MD1": "Gruppenphase - 1. Spieltag",
        "MD2": "Gruppenphase - 2. Spieltag",
        "MD3": "Gruppenphase - 3. Spieltag",        
        "MD4": "Achtelfinale",
        "MD5": "Viertelfinale",
        "MD6": "Halbfinale",
        "MD7": "Finale",
    }
    return matchday_mapping.get(matchday, matchday)

def get_end_result(id_to_find):
    row = df.loc[df['id_match'] == id_to_find]
    if row['winner_reason'].values[0] == "WIN_ON_PENALTIES":
        return f'{int(row['home_penalty'].values[0])} : {int(row['away_penalty'].values[0])}<br>(after Penalties)'
    else:
        return f'{int(row['home_score_total'].values[0])} : {int(row['away_score_total'].values[0])}'
        
df.loc[:, "marker_color"] = df["matchday_name"].apply(assign_marker_color)
df.loc[:, "matchday_name"] = df["matchday_name"].apply(
    assign_correct_matchday_names
)

map_stadiums.add_child(marker_cluster)

for index, record in df.iterrows():
    marker = folium.Marker(
        [record["stadium_latitude"], record["stadium_longitude"]],
        icon=folium.Icon(color="white", icon_color=record["marker_color"], shadow=True),
    )
    popup_text = f'Stadion:<br>{record["stadium_name"]}<br><br>Datum:<br>{record["date_time"].date()}<br><br>Anpfiff:<br>{record["date_time"].time()}<br><br>Zuschauerzahl<br>{record["match_attendance"]}<br><br>Spieltag:<br>{record["matchday_name"]}<br><br>Endergebnis:<br>{record["home_team"]} : {record["away_team"]}<br>{get_end_result(record["id_match"])}'
    marker.add_child(folium.Popup(popup_text, max_width=300))
    marker_cluster.add_child(marker)

map_stadiums
```
